```python
df["Investors Count"]= df["Investors Count"].fillna(df["Investors Count"].mean())
df.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
def outliers(dataframe,column):
    mean =dataframe[column].mean()
    std = dataframe[column].std()
    z_score= (dataframe[column]-mean)/std
    return  dataframe[np.abs(z_score)<3]
    
outliers(df,"Amount Normalized")
```

```python
categorical_column = ["Website Domain","Categories","Source Urls","Amount","Investors"]
for colm in categorical_column:
    df[colm]= le.fit_transform(df[colm].astype(str))
   
```

```python
df.head(10)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df["Amount Normalized"] = scaler.fit_transform(df["Amount Normalized"].values.reshape(-1,1))
df.head(10)
```
